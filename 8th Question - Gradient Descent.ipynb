```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1) Gradient function

```python
#derivative of f(x) = (x**4) + 4*(x**3) - 7*x**2 - 10*x
def grad_f(x):
   return 4*x*x*x + 12*x*x - 14*x - 10
```

# 2) 2D plot below

```python
def f(x):
   return (x**4) + 4*(x**3) - 7*x**2 - 10*x

#taking values from -10 to 10
x = np.linspace(-10, 10)

#plotting 2d graph
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.plot(x, f(x), color='red')
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show()
```

# 3) Python Code

```python
#initializing values

# Threshold
threshold = 0.001 #This tells us when to stop the algorithm

# Maximum iterations N_max
N_max = 10000 # maximum number of iterations
```

```python
def grad(init_x, learning_rate, threshold, N_max):
    #iterations variable
    N = 0
    #X is a list to hold the values of x at each step of gradient descent
    X = []
    #Y is a list to hold the values of grad_f(x) at each step of gradient descent
    Y = []
    X.append(init_x)
    Y.append(grad_f(init_x))
    flag=0
    
    while N < N_max:
        prev_x = init_x
        # x = x - rho(grad_f(x))
        init_x = init_x - learning_rate * grad_f(prev_x)
        X.append(init_x)
        Y.append(grad_f(prev_x))
        #iteration count
        N += 1 

        #Print iterations
        if abs(grad_f(prev_x)) <= threshold:
            print("converges at iteration", N, "with value", init_x, "\n")
            print("X values\n", X)
            flag=1
            break
            
    if(flag==0):        
        print("does not converge\n")
        
        print("X values\n", X)
        
    plt.plot(X, Y, color='red')
    plt.xlabel("X values")
    plt.ylabel("gradient of f(x)")
```

# 4) x=2 and learning rate=1/20

```python
# initial point x_0
init_x = 2

# learning rate
learning_rate = 1/20

grad(init_x, learning_rate, threshold, N_max)
```
